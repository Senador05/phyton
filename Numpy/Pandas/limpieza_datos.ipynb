```python
## No en este caso (los datos nulos solo son  207 frente a 20433 posiciones).
# # pero en caso de un numero de datos nulos elevados suele ser conveniente sustituir los datos nulos
#  por la media o la mediana y asi alterar menos la muestra que si los eliminamos.
```

```python
media = df['total_bedrooms'].mean()
print(media)
df['total_bedrooms'].fillna(media,inplace=True)
df.info()
df.describe()
```

```python
## o con la mediana

mediana = df['total_bedrooms'].median()
print(mediana)
df['total_bedrooms'].fillna(mediana,inplace=True)
df.info()
df.describe()
```

```python
import numpy as np

np.random.seed(0)
data = np.random.randint(low=0, high=11, size=1000)
data[0] = 100
data[1] = -100

data

```

```python
z_scores=(data-np.mean(data))/np.std(data)
z_scores
```

```python
umbral = 3
outliers = np.where(np.abs(z_scores) > umbral)
outliers
```

```python
data[outliers]
```

```python
# lo mismo para el df de las casas

z_scores=(df-np.mean(df))/np.std(df)
#print(z_scores)
umbral = 3
outliers = np.where(np.abs(z_scores) > umbral)
#print(outliers)

dfout_mask=~df.where(np.abs(z_scores) > umbral).isna()
dfout_mask
```

```python
q1 = np.percentile(data,25)
q3 = np.percentile(data,75)
iqr = q3 - q1
print(iqr)
print(np.mean(data))

umbra_sup = q3 + (1.5*iqr)
umbra_inf = q1 - (1.5*iqr)
print('umbrales:', umbra_inf, umbra_sup)

outliers = np.where((data < umbra_inf) | (data > umbra_sup))
outliers
data[outliers]
```
